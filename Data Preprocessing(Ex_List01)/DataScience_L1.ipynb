#### f) Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado.

```python
import pandas as pd

data1 = pd.read_csv('covid19br_vt1.csv', delimiter=';')

#checking the current types for each feature
for col in data1.columns:
    col_type = data1[col].dtype
    print("Column '{}' has type: {}".format(col, col_type))

#Based on previous EDA, assigning expected types for each feature
expected_types = {
    'id': float, ##checked
    'date': 'datetime64[ns]', ##checked 
    'hour': object, ##checked
    'ddi': float, ##checked
    'country': str, ##kinda too hard to check besides seeing if its a number, assuming it's all checked up
    'country_iso3': str, ##checked
    ##'ddd': object, ##checked
    'state': str, ##kinda too hard to check besides seeing if its a number, assuming it's all checked up
    'group': str,  ##kinda too hard to check, assuming it's all checked up
    'midia': bool, ##checked
    'url': bool ##checked
    ##'text': objectnot really needed to check, can be anything a dataframe cacn hold
}

wrong_types_count = 0

#hour format 'xx:xx'
def is_valid_hour(hour_str):
    try:
        pd.to_datetime(hour_str, format='%H:%M')
        return True
    except ValueError:
        return False

#date format 'xx/xx/xx'
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str, format='%d/%m/%y')
        return True
    except ValueError:
        return False


for col, expected_type in expected_types.items():
    if col == 'hour':
        invalid_hours = [val for val in data1[col] if not is_valid_hour(val)]

        if invalid_hours:
            wrong_types_count += len(invalid_hours)
            for invalid_hour in invalid_hours:
                print(f"Invalid hour format in column '{col}': {invalid_hour}")

    elif col == 'date':
        invalid_dates = [val for val in data1[col] if not is_valid_date(val)]
        if invalid_dates:
            wrong_types_count += len(invalid_dates)
            for invalid_date in invalid_dates:
                print(f"Invalid date format in column '{col}': {invalid_date}")

    else:
        if col not in data1.columns:
            print(f"Column '{col}' not found.")
            continue

        for i, val in enumerate(data1[col]):
            #convertion attempt
            try:
                converted_val = expected_type(val)
            except (ValueError, TypeError):
                # If conversion fails, increment wrong types counter
                wrong_types_count += 1
                print(f"Wrong type at row {i}, column '{col}': {val}")

# Print total count of wroung types
print(f"Total wrong types: {wrong_types_count}")
```

#### g) Crie uma coluna chamada “caracteres” contendo a quantidade de caracteres de coluna “text”.

```python
import pandas as pd

data1_1 = data1.assign(caracteres=data1['text'].str.len().fillna(0).astype(int))
data1_1.to_csv('data1_1.csv', index = True)

print(data1_1.head())
```

#### h) Crie uma coluna chamada “words” contendo a quantidade de palavras da coluna “text”.

```python
import pandas as pd

data1_2 = data1_1.assign(words=data1_1['text'].apply(lambda x: len(str(x).split()) if isinstance(x, str) else 0))
data1_2.to_csv('data1_2.csv', index = True)

print(data1_2.head())

print("-------------------------------------------info")

n_rows, n_col = data1_2.shape
print("Number of rows: ", n_rows)
print("Number of columns: ", n_col)
```

#### i) Crie uma coluna chamada “viral” contendo o valor 0 se o texto da mensagem (valor do campo “text”) não for encontrado em outras linhas e 1, caso contrário.

```python
import pandas as pd

data1_3 = data1_2.assign(viral=data1_2['text'].duplicated(keep=False)*1)
data1_3.to_csv('data1_3.csv', index = True)

print(data1_3.head())

print("-------------------------------------------info")

n_rows, n_col = data1_3.shape
print("Number of rows: ", n_rows)
print("Number of columns: ", n_col)
```

#### j) Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto armazenado no atributo “text” aparece no dataset.

```python
import pandas as pd

data1_4 = data1_3.assign(sharings=data1_3['text'].map(data1_3['text'].value_counts()).fillna(0).astype(int))
data1_4.to_csv('data1_4.csv', index = True)

print(data1_4.head())

print("-------------------------------------------info")

n_rows, n_col = data1_4.shape
print("Number of rows: ", n_rows)
print("Number of columns: ", n_col)
```

#### k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos

```python
from textblob import TextBlob
from googletrans import Translator
import pandas as pd

#gets the polarity atribute from a pt_text
def get_polarity(text_pt, n):

    global texts_count
    
    if not isinstance(text_pt, str):
        return None
    
    try:

        if texts_count >= n:
            return None
        texts_count += 1

        if len(str(text_pt)) > 430:
            ##print(f"Skipping text with length {len(text_pt)} characters: {text_pt}")
            return -1
        
        #translates the text so it can be used on textblob
        translator = Translator()
        text_en = translator.translate(text_pt, src='pt', dest='en').text

        polarity = TextBlob(text_en).sentiment.polarity

        if polarity <= -0.5:
            return -1
        elif -0.5 < polarity < 0.5:
            return 0
        else:
            return 1
    
    except Exception as e:
        print(f"Error occurred during translation: {e}")
        return None


texts_count = 0
n = 100

data1_5 = data1_4.head(n).assign(sentiment=data1_4['text'].head(n).apply(get_polarity, n=n))
data1_5.to_csv('data1_5.csv', index = True)
print(data1_5.head())
```

#### m) Identificar inconsistências entre os atributos (features).

```python
import pandas as pd

print(data1_6.dtypes)
print("-------------------------------------------------------------------")
print(data1_6.describe())
print("-------------------------------------------------------------------")

data1 = pd.read_csv('covid19br_vt1.csv', delimiter=';')

# Checking the current types of each feature
for col in data1.columns:
    col_type = data1[col].dtype
    print("Column '{}' has type: {}".format(col, col_type))

# Based on previous EDA, assigning expected types for each feature
expected_types = {
    'id': float, 
    'date': 'datetime64[ns]', 
    'hour': object, 
    'ddi': float, 
    'country': str,
    'country_iso3': str,
    'ddd': str,  # Changing the expected type of 'ddd' to string
    'state': str,
    'group': str,
    'midia': bool,
    'url': bool
}

# Initialize counter for wrong types
wrong_types_count = 0

# Function to check if a string is a valid date
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str, format='%d/%m/%y')  # Changed the format string to '%d/%m/%y'
        return True
    except ValueError:
        return False
    
def is_valid_hour(hour_str):
    try:
        pd.to_datetime(hour_str, format='%H:%M')
        return True
    except ValueError:
        return False

def is_valid_country_iso3(iso3_str):
    return len(iso3_str) == 3 if pd.notna(iso3_str) and isinstance(iso3_str, str) else True

def is_valid_ddd(ddd_str):
    return len(ddd_str) <= 2 if pd.notna(ddd_str) and isinstance(ddd_str, str) else True

# Iterate over each column
for col, expected_type in expected_types.items():
    # Special handling for hour column
    if col == 'hour':
        # Check if all values in the hour column are in the format 'xx:xx'
        invalid_hours = [val for val in data1[col] if not is_valid_hour(val)]
        if invalid_hours:
            wrong_types_count += len(invalid_hours)
            for invalid_hour in invalid_hours:
                print(f"Inconsistency at '{col}': {invalid_hour}")
    elif col == 'date':
        # Check if all values in the date column are in the correct format
        invalid_dates = [val for val in data1[col] if not is_valid_date(val)]
        if invalid_dates:
            wrong_types_count += len(invalid_dates)
            for invalid_date in invalid_dates:
                print(f"Inconsistency at '{col}': {invalid_date}")
    elif col == 'country_iso3':
        # Check if all values in the country_iso3 column have exactly three letters
        invalid_iso3 = [iso3 for iso3 in data1[col] if not is_valid_country_iso3(iso3)]
        if invalid_iso3:
            wrong_types_count += len(invalid_iso3)
            for invalid_value in invalid_iso3:
                print(f"Inconsistency at '{col}': {invalid_value}")
    elif col == 'ddd':
        # Check if all values in the ddd column have exactly two digits
        invalid_ddd = [ddd for ddd in data1[col] if not is_valid_ddd(ddd)]
        if invalid_ddd:
            wrong_types_count += len(invalid_ddd)
            for invalid_value in invalid_ddd:
                print(f"Inconsistency at '{col}': {invalid_value}")
    else:
        # Check if the column exists
        if col not in data1.columns:
            print(f"Column '{col}' not found.")
            continue

        # Iterate over each cell in the column
        for i, val in enumerate(data1[col]):
            # Attempt to convert the value to the expected type
            try:
                converted_val = expected_type(val)
            except (ValueError, TypeError):
                # If conversion fails, increment wrong types counter
                wrong_types_count += 1
                print(f"Inconsistency at {i}, column '{col}': {val}")

# Print total count of wrong types
print(f"Inconsistencies: {wrong_types_count}")

```
